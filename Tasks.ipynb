```python
#Write a Python script that creates a NumPy array with random integers and performs basic mathematical operations like addition, subtraction, multiplication, and division on it.
```

```python
import numpy as np

# Create a random array of integers
array_size = 5  # Change this to the desired size of your array
min_value = 1   # Change this to the minimum value for random integers
max_value = 10  # Change this to the maximum value for random integers
random_array = np.random.randint(min_value, max_value + 1, size=array_size)

# Display the random array
print("Random Array:")
print(random_array)

# Perform basic mathematical operations
scalar = 2  # Change this to the desired scalar value for operations

# Addition
addition_result = random_array + scalar
print("\nAddition Result:")
print(addition_result)

# Subtraction
subtraction_result = random_array - scalar
print("\nSubtraction Result:")
print(subtraction_result)

# Multiplication
multiplication_result = random_array * scalar
print("\nMultiplication Result:")
print(multiplication_result)

# Division (Note: Watch out for division by zero)
divisor = 3  # Change this to the desired divisor
division_result = random_array / divisor
print("\nDivision Result:")
print(division_result)
```

```python
#Using Pandas, load a dataset of your choice (you can find datasets online or use your own). Clean the data and perform basic data analysis tasks like calculating mean, median, and mode for specific columns.
```

```python
import pandas as pd
import seaborn as sns

# Load the Iris dataset from Seaborn
iris_data = sns.load_dataset('iris')

# Display the first few rows of the dataset
print("Sample of the Iris dataset:")
print(iris_data.head())

# Check for missing values
missing_values = iris_data.isnull().sum()
print("\nMissing Values:")
print(missing_values)

# Calculate mean, median, and mode for the 'sepal_length' column
sepal_length_mean = iris_data['sepal_length'].mean()
sepal_length_median = iris_data['sepal_length'].median()
sepal_length_mode = iris_data['sepal_length'].mode().iloc[0]  # Get the first mode if multiple modes exist

# Display the calculated statistics
print("\nMean of Sepal Length:", sepal_length_mean)
print("Median of Sepal Length:", sepal_length_median)
print("Mode of Sepal Length:", sepal_length_mode)
```

```python
#Create two NumPy arrays and perform operations to concatenate them vertically and horizontally.
```

```python
import numpy as np

# Create two sample NumPy arrays
array1 = np.array([[1, 2, 3],
                   [4, 5, 6]])

array2 = np.array([[7, 8, 9],
                   [10, 11, 12]])

# Vertically concatenate the arrays
vertical_concatenation = np.concatenate((array1, array2), axis=0)

# Horizontally concatenate the arrays
horizontal_concatenation = np.concatenate((array1, array2), axis=1)

# Display the original arrays and concatenated arrays
print("Array 1:")
print(array1)

print("\nArray 2:")
print(array2)

print("\nVertically Concatenated Array:")
print(vertical_concatenation)

print("\nHorizontally Concatenated Array:")
print(horizontal_concatenation)
```

```python
#Given a dataset (you can choose one or use your own), use Pandas to filter and extract rows that meet specific criteria (e.g., filtering data for a specific date range).
```

```python
import pandas as pd

# Sample dataset: Daily stock prices
data = {
    'Date': ['2023-01-01', '2023-01-02', '2023-01-03', '2023-01-04', '2023-01-05'],
    'Ticker': ['AAPL', 'AAPL', 'AAPL', 'AAPL', 'AAPL'],
    'Open': [150.0, 152.5, 153.0, 155.0, 156.0],
    'Close': [151.0, 153.0, 153.5, 155.5, 157.0],
}

# Create a DataFrame
df = pd.DataFrame(data)

# Convert the 'Date' column to a datetime type
df['Date'] = pd.to_datetime(df['Date'])

# Define the date range for filtering
start_date = '2023-01-02'
end_date = '2023-01-04'

# Filter the data for the specified date range
filtered_data = df[(df['Date'] >= start_date) & (df['Date'] <= end_date)]

# Display the filtered data
print("Filtered Data:")
print(filtered_data)
```

```python
#Create a line plot or scatter plot using Matplotlib and customize it by adding a legend, different marker styles, and changing the line colors.
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Sample data
x = np.array([1, 2, 3, 4, 5])
y1 = x * 2
y2 = x * x
y3 = 10 - x

# Create the scatter plot
plt.scatter(x, y1, label='2x', marker='o', color='blue')
plt.scatter(x, y2, label='x^2', marker='s', color='green')
plt.scatter(x, y3, label='10-x', marker='^', color='red')

# Add a legend
plt.legend()

# Customize the plot
plt.title('Customized Scatter Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')

# Show the plot
plt.grid(True)
plt.show()
```

```python
#Use NumPy to calculate descriptive statistics like variance, standard deviation, and correlation coefficients for a dataset of your choice (you can find datasets online or use your own).
```

```python
import numpy as np

# Create a sample dataset
X = np.array([1, 2, 3, 4, 5])
Y = np.array([2, 4, 5, 4, 5])

# Calculate variance and standard deviation for X and Y
variance_X = np.var(X)
std_deviation_X = np.std(X)
variance_Y = np.var(Y)
std_deviation_Y = np.std(Y)

print("Variance of X:", variance_X)
print("Standard Deviation of X:", std_deviation_X)
print("Variance of Y:", variance_Y)
print("Standard Deviation of Y:", std_deviation_Y)

# Calculate the correlation coefficient between X and Y
correlation_coefficient = np.corrcoef(X, Y)[0, 1]

print("\nCorrelation Coefficient between X and Y:", correlation_coefficient)
```

```python
#Given a dataset (you can choose one or use your own), use Pandas to group data by a specific column and calculate summary statistics (e.g., mean, median) for each group.
```

```python
import pandas as pd
import seaborn as sns

# Load the Titanic dataset from Seaborn
titanic_data = sns.load_dataset('titanic')

# Display the first few rows of the dataset
print("Sample of the Titanic dataset:")
print(titanic_data.head())

# Group passengers by class and calculate mean and median age for each class
grouped_data = titanic_data.groupby('class')['age'].agg(['mean', 'median'])

# Display the summary statistics
print("\nSummary statistics of age by class:")
print(grouped_data)
```

```python
#Create a Matplotlib figure with multiple subplots to display different aspects of a dataset or related datasets (you can choose the dataset and the type of plots).
```

```python
import matplotlib.pyplot as plt
import seaborn as sns  # Import Seaborn for dataset loading

# Load the Iris dataset
iris = sns.load_dataset("iris")

# Create a figure with multiple subplots
fig, axes = plt.subplots(2, 2, figsize=(12, 10))

# Subplot 1: Histogram for Sepal Length
sns.histplot(iris["sepal_length"], bins=10, kde=True, ax=axes[0, 0])
axes[0, 0].set_title("Histogram of Sepal Length")
axes[0, 0].set_xlabel("Sepal Length")
axes[0, 0].set_ylabel("Frequency")

# Subplot 2: Scatter Plot for Sepal Length vs. Sepal Width
sns.scatterplot(data=iris, x="sepal_length", y="sepal_width", hue="species", ax=axes[0, 1])
axes[0, 1].set_title("Scatter Plot of Sepal Length vs. Sepal Width")
axes[0, 1].set_xlabel("Sepal Length")
axes[0, 1].set_ylabel("Sepal Width")

# Subplot 3: Scatter Plot for Petal Length vs. Petal Width
sns.scatterplot(data=iris, x="petal_length", y="petal_width", hue="species", ax=axes[1, 0])
axes[1, 0].set_title("Scatter Plot of Petal Length vs. Petal Width")
axes[1, 0].set_xlabel("Petal Length")
axes[1, 0].set_ylabel("Petal Width")

# Subplot 4: Box Plot for Sepal Length by Species
sns.boxplot(data=iris, x="species", y="sepal_length", ax=axes[1, 1])
axes[1, 1].set_title("Box Plot of Sepal Length by Species")
axes[1, 1].set_xlabel("Species")
axes[1, 1].set_ylabel("Sepal Length")

# Adjust layout and show the figure
plt.tight_layout()
plt.show()
```

```python
#visualizations (e.g., bar charts, pie charts, scatter plots) using Pandas to represent different aspects of the data, such as trends or distributions.
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Sample movie dataset
data = {
    'Movie': ['Movie A', 'Movie B', 'Movie C', 'Movie D', 'Movie E'],
    'Year': [2010, 2015, 2017, 2012, 2019],
    'BoxOffice (Millions USD)': [100, 150, 80, 120, 200],
    'Genre': ['Action', 'Comedy', 'Drama', 'Adventure', 'Sci-Fi']
}

# Create a DataFrame
df = pd.DataFrame(data)

# Visualization 1: Bar chart of average box office earnings by genre
average_earnings_by_genre = df.groupby('Genre')['BoxOffice (Millions USD)'].mean()
average_earnings_by_genre.plot(kind='bar', color='skyblue')
plt.title('Average Box Office Earnings by Genre')
plt.xlabel('Genre')
plt.ylabel('Average Earnings (Millions USD)')
plt.xticks(rotation=45)
plt.show()

# Visualization 2: Pie chart of the number of movies by genre
movie_count_by_genre = df['Genre'].value_counts()
movie_count_by_genre.plot(kind='pie', autopct='%1.1f%%', startangle=90, colors=['lightcoral', 'lightgreen', 'lightblue', 'lightyellow', 'lightpink'])
plt.title('Number of Movies by Genre')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()

# Visualization 3: Scatter plot of box office earnings vs. year
plt.scatter(df['Year'], df['BoxOffice (Millions USD)'], c='gold', s=100, alpha=0.7)
plt.title('Box Office Earnings vs. Year')
plt.xlabel('Year')
plt.ylabel('Box Office Earnings (Millions USD)')
plt.grid(True)
plt.show()
```
